## Observations and Insights

1- Capomulin is the most used treatmet (230 times)

2- Capomulin and Ramicane have the most densed results due to their low variance and standard deviation.

3- Almost the half population of tested mice are men and the other half women (50.6% men and 49.4% women)

4- Although the Infubinol regimen has shown promising results, it contains one outlier result.

5- The tumor volume in some cases is bigger at the end of the treatment than durarion of the treatment.

6- There is a positive relation between mouse weight and average tumor volume for Capomulin treatment

```python
# Dependencies and Setup

import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
```

```python
# Read and Save study data files

mouse_metadata = "Mouse_metadata.csv"
study_results = "Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)

# Combine the data into a single dataset

combined_data = pd.merge(mouse_metadata, study_results, on = "Mouse ID", how = "outer")

combined_data.head()
```

## Summary statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

tumor_volume_data_groupby_regimen = combined_data[["Drug Regimen","Tumor Volume (mm3)"]].groupby("Drug Regimen")
tumor_volume_count = tumor_volume_data_groupby_regimen["Tumor Volume (mm3)"].count()
tumor_volume_mean = tumor_volume_data_groupby_regimen["Tumor Volume (mm3)"].mean()
tumor_volume_mean = round(tumor_volume_mean,2)
tumor_volume_median = tumor_volume_data_groupby_regimen["Tumor Volume (mm3)"].median()
tumor_volume_median = round(tumor_volume_median,2)
tumor_volume_var = tumor_volume_data_groupby_regimen["Tumor Volume (mm3)"].var()
tumor_volume_var = round(tumor_volume_var,2)
tumor_volume_std = tumor_volume_data_groupby_regimen["Tumor Volume (mm3)"].std()
tumor_volume_std = round(tumor_volume_std,2)
tumor_volume_sem = tumor_volume_data_groupby_regimen["Tumor Volume (mm3)"].sem()
tumor_volume_sem = round(tumor_volume_sem,2)

statistic_info = pd.DataFrame({"Count":tumor_volume_count,
                               "Mean":tumor_volume_mean,
                               "Median":tumor_volume_median,
                               "Variance":tumor_volume_var,
                               "STD":tumor_volume_std,
                               "SEM":tumor_volume_sem})

# table format

statistic_info.style.set_caption('Data of Tumor Volume (mm3)').set_table_styles([{
'selector': 'caption',
'props': [
('color', 'blue'),
('font-size', '15px'),
('font-weight', 'bold'),
('text-align', 'center')
]
}]).format({"Mean": "{:}",
           "Median":"{:}",
           "Variance" : "{:}",
           "STD": "{:}",
           "SEM" : "{:}"})


```

## Quartiles, outliers and boxplots

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens. Calculate the IQR and quantitatively determine if there are any potential outliers.

# FYI:
# FTVM = final_tumor_volume_of_each_mouse
# FTVM_Name = final_tumor_volume_of_each_mouse_across_Name

FTVM =  combined_data[["Mouse ID","Drug Regimen","Tumor Volume (mm3)"]]\
        .groupby(["Drug Regimen", "Mouse ID" ]).last()\
        .reset_index()
            

FTVM_Capomulin = FTVM.loc[(FTVM["Drug Regimen"] == "Capomulin"),:]
FTVM_Ramicane = FTVM.loc[(FTVM["Drug Regimen"] == "Ramicane"),:]
FTVM_Infubinol = FTVM.loc[(FTVM["Drug Regimen"] == "Infubinol"),:]
FTVM_Ceftamin = FTVM.loc[(FTVM["Drug Regimen"] == "Ceftamin"),:]


quantile_Capomulin = FTVM_Capomulin["Tumor Volume (mm3)"].quantile([0.25,0.5,0.75])
quantile_Ramicane = FTVM_Ramicane["Tumor Volume (mm3)"].quantile([0.25,0.5,0.75])
quantile_Infubinol = FTVM_Infubinol["Tumor Volume (mm3)"].quantile([0.25,0.5,0.75])
quantile_Ceftamin = FTVM_Ceftamin["Tumor Volume (mm3)"].quantile([0.25,0.5,0.75])

lowerq_Capomulin = quantile_Capomulin[0.25]
lowerq_Ramicane = quantile_Ramicane[0.25]
lowerq_Infubinol = quantile_Infubinol[0.25]
lowerq_Ceftamin = quantile_Ceftamin[0.25]

upperq_Capomulin = quantile_Capomulin[0.75]
upperq_Ramicane = quantile_Ramicane[0.75]
upperq_Infubinol = quantile_Infubinol[0.75]
upperq_Ceftamin = quantile_Ceftamin[0.75]

iqr_Capomulin = upperq_Capomulin-lowerq_Capomulin
iqr_Ramicane = upperq_Ramicane-lowerq_Ramicane
iqr_Infubinol = upperq_Infubinol-lowerq_Infubinol
iqr_Ceftamin = upperq_Ceftamin-lowerq_Ceftamin

lowerband_Capomulin = lowerq_Capomulin-(1.5*iqr_Capomulin)
lowerband_Ramicane = lowerq_Ramicane-(1.5*iqr_Ramicane)
lowerband_Infubinol = lowerq_Infubinol-(1.5*iqr_Infubinol)
lowerband_Ceftamin = lowerq_Ceftamin-(1.5*iqr_Ceftamin)

upperband_Capomulin = upperq_Capomulin+(1.5*iqr_Capomulin)
upperband_Ramicane = upperq_Ramicane+(1.5*iqr_Ramicane)
upperband_Infubinol = upperq_Infubinol+(1.5*iqr_Infubinol)
upperband_Ceftamin = upperq_Ceftamin+(1.5*iqr_Ceftamin)

potential_outliers_Capomulin = FTVM_Capomulin.loc[(FTVM_Capomulin["Tumor Volume (mm3)"] < 
                               lowerband_Capomulin)|(FTVM_Capomulin["Tumor Volume (mm3)"] > 
                               upperband_Capomulin),:]

potential_outliers_Ramicane = FTVM_Ramicane.loc[(FTVM_Ramicane["Tumor Volume (mm3)"] < 
                              lowerband_Ramicane)|(FTVM_Ramicane["Tumor Volume (mm3)"] > 
                              upperband_Ramicane),:]

potential_outliers_Infubinol = FTVM_Infubinol.loc[(FTVM_Infubinol["Tumor Volume (mm3)"] < 
                               lowerband_Infubinol)|(FTVM_Infubinol["Tumor Volume (mm3)"] > 
                               upperband_Infubinol),:]

potential_outliers_Ceftamin = FTVM_Ceftamin.loc[(FTVM_Ceftamin["Tumor Volume (mm3)"] < 
                              lowerband_Ceftamin)|(FTVM_Ceftamin["Tumor Volume (mm3)"] > 
                              upperband_Ceftamin),:]

print('Capomulin dataset has '+str(len(potential_outliers_Capomulin))+" Outliers.")
if not(potential_outliers_Capomulin.empty):
    print("Oulier/s is/are"+potential_outliers_Capomulin.to_string(index=False))
    
print('Ramicane dataset has '+str(len(potential_outliers_Ramicane))+" Outliers.")
if not(potential_outliers_Ramicane.empty):
    print("Oulier/s is/are"+potential_outliers_Ramicane.to_string(index=False))
    
print('Infubinol dataset has '+str(len(potential_outliers_Infubinol))+" Outliers.")
if not(potential_outliers_Infubinol.empty):
    print("Oulier/s is/are\n"+potential_outliers_Infubinol.to_string(index=False))

print('Ceftamin dataset has '+str(len(potential_outliers_Ceftamin))+" Outliers.")
if not(potential_outliers_Ceftamin.empty):
    print("Oulier/s is/are\n"+potential_outliers_Ceftamin.to_string(index=False))
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

data = [FTVM_Capomulin["Tumor Volume (mm3)"],FTVM_Ramicane["Tumor Volume (mm3)"],\
        FTVM_Infubinol["Tumor Volume (mm3)"],FTVM_Ceftamin["Tumor Volume (mm3)"]]

fig, ax = plt.subplots(figsize=(8, 8))
fig.suptitle("Final Tumor Volume of Each Mouse Across Four Promissing Regimens",fontsize = 14)

A = ax.boxplot(data, patch_artist=True, sym = 'm*')
colors = ["red","green","blue","purple"]

for box,c in zip(A['boxes'],colors):
    box.set(color = c, linewidth=2)
    
ax.set_ylabel("Final Tumor Volume(mm3)",fontsize =15)

ax.text(0.75, max(FTVM_Capomulin["Tumor Volume (mm3)"])+1, "Capomulin", transform=ax.transData)
ax.text(1.75, max(FTVM_Ramicane["Tumor Volume (mm3)"])+1, "Ramicane", transform=ax.transData)
ax.text(2.78, max(FTVM_Infubinol["Tumor Volume (mm3)"])+1, "Infubinol", transform=ax.transData)
ax.text(3.75, max(FTVM_Ceftamin["Tumor Volume (mm3)"])+1, "Ceftamin", transform=ax.transData)
ax.text(2.85, 29, "Outlier", transform=ax.transData)
ax.annotate("", xy=(3, 35.5), xytext=(3, 30.5),
            arrowprops=dict(arrowstyle="->"))

plt.show()
```

## Line and scatter plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin

t565_data = combined_data.loc[combined_data["Mouse ID"]=="t565",
                              ["Drug Regimen","Timepoint", "Tumor Volume (mm3)"]]

x_value = t565_data["Timepoint"]
y_value = t565_data["Tumor Volume (mm3)"]
plt.plot(x_value,y_value,color = "r", marker = "o")


plt.xlim(min(x_value),max(x_value))
plt.ylim(30,50)
plt.xlabel("Time Point")
plt.ylabel("Tumor Volume(mm3)")
plt.title("Tumor Volume vs Time Point of the t565 Mouse Treated with Capomulin ")

plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen

capomulin_regimen_groupedby_mouseID = combined_data.loc[combined_data["Drug Regimen"]=="Capomulin",
                                      ["Mouse ID","Weight (g)","Tumor Volume (mm3)"]]\
                                      .groupby("Mouse ID").mean()

average_tumor_volume = capomulin_regimen_groupedby_mouseID["Tumor Volume (mm3)"]
mouse_weight = capomulin_regimen_groupedby_mouseID["Weight (g)"]

plt.scatter(mouse_weight, average_tumor_volume, c = mouse_weight)
plt.ylabel("Average Tumor Volume")
plt.xlabel("Mouse Weight")
plt.title("Capomulin Regimen")
plt.show()
```

```python
# Calculate the correlation coefficient and linear regression model for mouse weight and average tumor volume for the Capomulin regimen

(slope, intercept,rvalue,pvalue,stdrr) = st.linregress( mouse_weight,average_tumor_volume)


plt.scatter( mouse_weight,average_tumor_volume, c = mouse_weight)
plt.ylabel("Average Tumor Volume")
plt.xlabel("Mouse Weight")
plt.title("Capomulin Regimen")
plt.xlim(min(mouse_weight),max(mouse_weight))


equation = slope*mouse_weight+intercept
plt.plot(mouse_weight, equation, color ="r")

equation = "y="+str(round(slope,2))+"*x+"+str(round(intercept,2))
plt.annotate(equation, xy=(18, 38), fontsize=12, color ="r")



plt.show()
```
